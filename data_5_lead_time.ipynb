````python
df = pd.read_csv('train.csv')
df.shape
df['booking_status'].head()
booking_status = {
    "Canceled": 1,
    "Not_Canceled": 0
}
df['booking_status'] = df['booking_status'].map(booking_status)
df['booking_status'].head()
````

````python
df = df.drop(["arrival_year", "arrival_month", "arrival_date"], axis=1)
````

````python
meal_plan = {"Meal Plan 1": 0, "Not Selected": 1, "Meal Plan 2": 2, "Meal Plan 3": 3}
df['type_of_meal_plan'] = df['type_of_meal_plan'].map(meal_plan)

room_type = {
    "Room_Type 1": 0,
    "Room_Type 4": 1,
    "Room_Type 6": 3,
    "Room_Type 2": 2,
    "Room_Type 5": 4,
    "Room_Type 7": 5,
    "Room_Type 3": 6
}
df['room_type_reserved'] = df['room_type_reserved'].map(room_type)

market_segment = {
    "Online": 1,
    "Offline": 0,
    "Corporate": 2,
    "Complementary": 4,
    "Aviation": 3
}
df['market_segment_type'] = df['market_segment_type'].map(market_segment)

category_cols = ["type_of_meal_plan", "required_car_parking_space", "room_type_reserved", "market_segment_type", "repeated_guest"]
for col in category_cols:
    display(df[col].value_counts())
    print("#" * 50)
````

````python
df["market_segment_type"].replace({3:4}).value_counts()
````

````python
category_cols = ["type_of_meal_plan", "required_car_parking_space", "room_type_reserved", "market_segment_type", "repeated_guest"]
X_log_1_dummy = []
for col in category_cols:
    if col == "type_of_meal_plan":
        col_values = df[col].replace(3, 2)
    elif col == "room_type_reserved":
        col_values = df[col].replace({6:4, 5:4})
    elif col == "market_segment_type":
        col_values = df[col].replace({3:4})
    else:
        col_values = df[col]
    X_log_1_dummy.append(
        pd.get_dummies(col_values, prefix=col, prefix_sep='__', drop_first=True)
    )
X_log_1_dummy = pd.concat(X_log_1_dummy, axis=1)
df = pd.concat([df.drop(category_cols, axis=1), X_log_1_dummy], axis=1)
````

````python
df.keys()
````

````python
df = pd.DataFrame(df)
df["lead_time"] = np.log1p(df["lead_time"])
df["avg_price_per_room/"] = df["avg_price_per_room"].apply(lambda x: 209 if x>209 else x) # 나머지 99% 외의 나머지는 209로

df["lead_time__cut"] = pd.cut(df["lead_time"], 15, labels=[f"lead_time__{i}" for i in range(1,16)])
df["avg_price_per_room__cut"] = pd.cut(df["avg_price_per_room"], 11, labels=[f"avg_price_per_room__{i}" for i in range(1,12)])

df_dummy = [
    pd.get_dummies(df["lead_time__cut"], prefix="lead_time", prefix_sep='__', drop_first=True),
    pd.get_dummies(df["avg_price_per_room__cut"], prefix="avg_price_per_room", prefix_sep='__', drop_first=True)
]
df = pd.concat([df.drop(["Booking_ID", "lead_time", "lead_time__cut", "avg_price_per_room__cut"], axis=1)] + df_dummy, axis=1)
df.keys()
````

````python
from datetime import timedelta

data = pd.read_csv('train.csv')


for data in [train, test]:
    df = data.copy()
    df['arrival_year_month'] = pd.to_datetime(df['arrival_year'].astype(str) + df['arrival_month'].astype(str), format='%Y%m')

    df.loc[df.arrival_date > df.arrival_year_month.dt.days_in_month, 'arrival_date'] = (
        df.loc[df.arrival_date > df.arrival_year_month.dt.days_in_month, 'arrival_year_month'].dt.days_in_month   
    )
    df = df.drop('arrival_year_month', axis=1)

    data['nights'] = data['no_of_weekend_nights'] + data['no_of_week_nights']
    data['lead_x_nights'] = data['lead_time'] * data['nights']
    data['arrival'] = pd.to_datetime(df['arrival_year'].astype(str) + "/" + df['arrival_month'].astype(str) + "/" + df['arrival_date'].astype(str))
    data['booking'] = data['arrival'] - pd.to_timedelta(data['lead_time'], unit='D')
    data['departure'] = data['arrival'] + pd.to_timedelta(data['nights'], unit='D')
    
for df in [train, test]:
    df.drop(['booking', 'arrival', 'departure'], axis=1, inplace=True)
features = test.columns.to_list()

train['no_of_adults'] += train['no_of_children']
test['no_of_adults'] += test['no_of_children']

train = train.drop(['no_of_children'], axis=1)
test = test.drop(['no_of_children'], axis=1) 
````

````python
from sklearn.preprocessing import StandardScaler

features = [
    'no_of_adults',
    'no_of_weekend_nights',
    'no_of_week_nights',
    'lead_time',
    'arrival_month',
    'arrival_date',
    'no_of_previous_cancellations',
    'no_of_previous_bookings_not_canceled',
    'avg_price_per_room',
    'no_of_special_requests'
]
scaler = StandardScaler()
train[features] = scaler.fit_transform(train[features])
test[features] = scaler.transform(test[features])
````

````python
train = pd.get_dummies(data=train, columns=categoricals)
test = pd.get_dummies(data=test, columns=categoricals)
````
